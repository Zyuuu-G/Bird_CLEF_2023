```python
ipd.Audio(audio_file0[0])
```

```python
# read/load audio
data, sr = librosa.load(audio_file0[0])
print(data.shape)
print(sr)
# plot the "sound"
pd.Series(data).plot(figsize=(15, 5),lw=1,title='raw audio')

plt.show()
```

```python
# short time Fouier transform and spectrogram
D = librosa.stft(data)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.min)
S_db.shape


# Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,x_axis='time',y_axis='log',ax=ax)
ax.set_title('Spectrogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
C = librosa.feature.chroma_stft(y=data,sr=sr)

# Plot chromagram
plt.figure(figsize=(10, 5))
librosa.display.specshow(C, x_axis='time', y_axis='chroma')
plt.colorbar()
plt.title('Chromagram ')
plt.tight_layout()
plt.show()
```
